```python
import pandas as pd

paths = {
    "KF":        "../backend/data/kf_outputs/kf_fair_eval_hives.csv",
    "EnKF":      "../backend/data/enkf_outputs/enkf_fair_eval_hives.csv",
    "EKF":       "../backend/data/ekf_outputs_fair/ekf_eval_hives.csv",
    "AdaptiveR": "../backend/data/kf_outputs_experiment_adaptiveR/kf_eval_hives_adaptiveR_GATED_FAIR.csv",
}

sets = {k: set(pd.read_csv(v)["hive_id"].astype(int)) for k, v in paths.items()}

base_name = list(sets.keys())[0]
base = sets[base_name]

for name, s in sets.items():
    print(name, len(s), "match?" , s == base)
    if s != base:
        print("only in base:", sorted(base - s))
        print("only in", name, ":", sorted(s - base))
```
